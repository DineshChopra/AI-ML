```
import tensorflow as tf
```

```
import numpy as np
```

```
features_celsius = np.array([-40, -10,  0,  8, 15, 22,  38], dtype=float)
labels_fahrenheit = np.array([-40,  14, 32, 46, 59, 72, 100], dtype=float)
```

```
layer0 = tf.keras.layers.Dense(units = 10, input_shape = [1])
layer1 = tf.keras.layers.Dense(units = 10)
layer2 = tf.keras.layers.Dense(units = 1)

model = tf.keras.Sequential([layer0, layer1, layer2]);
```

```
# loss = 'mean_squared_error'
# optimizer = tf.keras.optimizers.Adam(0.1)
model.compile(loss = 'mean_squared_error', optimizer = tf.keras.optimizers.Adam(0.1))
```

```
history = model.fit(features_celsius, labels_fahrenheit, epochs=500, verbose=False)
```

```
print(model.predict([100.0]))
```

```
import matplotlib.pyplot as plt
plt.xlabel('Epoch Number')
plt.ylabel('Loss Magnitude')
plt.plot(history.history['loss'])
```
